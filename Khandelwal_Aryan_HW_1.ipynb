```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

### Load data

```python
# Read data from file (breast_cancer_diagnosis.csv) into dataframe
breast_cancer_diagnosis = pd.read_csv('data/breast_cancer_diagnosis.csv')
```

### Handle missing values
NOTES:
* You cannot drop the column with missing values
* You cannot drop the rows with missing values
* You cannot replace the missing values with 0  

Remember to use "inplace=True"

```python
# Figure out strategy to replace values where radius is NaN
breast_cancer_diagnosis[breast_cancer_diagnosis['radius'].isnull()]
breast_cancer_diagnosis.fillna(breast_cancer_diagnosis['radius'].mean())
```

```python
# Replace values where radius is NaN based on above strategy
breast_cancer_diagnosis_new = breast_cancer_diagnosis.fillna(breast_cancer_diagnosis['radius'].mean())
```

### Check for missing values again

```python
breast_cancer_diagnosis_new.isnull().sum()
```

### Drop non-numeric variables
Remember to use "inplace=True"

```python
non_numeric_columns = breast_cancer_diagnosis_new.select_dtypes(exclude='number')
non_numeric_columns.columns
breast_cancer_diagnosis_new.drop(non_numeric_columns, axis=1, inplace=True)
breast_cancer_diagnosis_new.shape
```

### Review updated dataframe

```python
# Display first 10 rows
breast_cancer_diagnosis_new.head(10)
```

### Separate independent and dependent variables
* Independent variables: All remaining variables except Diagnosis
* Dependent variable: Diagnosis

```python
# Prepare dataset for model training
X = breast_cancer_diagnosis_new.iloc[:, :-1]   
y = breast_cancer_diagnosis_new.iloc[:, -1]   
```

### Split data into training and test sets

```python
# Split data into training data and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=50)
```

### Train LogisticRegression model

```python
# Train model
model = LogisticRegression(max_iter = 10000)
model.fit(X_train, y_train)
```

### Test model

```python
# Generate predictions against the test set
predictions = model.predict(X_test)
```

### Model evaluation

```python
# Print classification report
print(classification_report(y_test, predictions))
```

```python
# Print model accuracy
accuracy = model.score(X_test, y_test)
print("accuracy =", round((accuracy * 100), 2), "%")
```
